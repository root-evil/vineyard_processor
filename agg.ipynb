```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from skimage import io

tiff_name = 'data/KRA_TAVG_100m.tif'
img = io.imread(tiff_name)
```

```python
nan_val = -2147483647
img_mask_nan_val = img == nan_val
```

```python
img_agr_stat = img.mean(axis = 2)
```

```python
io.imsave('agregation/mean_TAVG_100m.tif', img_agr_stat, plugin='tifffile')
```

```python
tiff_name = 'data/KRA_TMAX_100m.tif'
img = io.imread(tiff_name)
img_agr_stat = img.max(axis = 2)
io.imsave('agregation/max_TAVG_100m.tif', img_agr_stat, plugin='tifffile')
```

```python
tiff_name = 'data/KRA_TMAX_100m.tif'
img = io.imread(tiff_name)
img_agr_stat = img.min(axis = 2)
io.imsave('agregation/min_TAVG_100m.tif', img_agr_stat, plugin='tifffile')
```

```python
tiff_name = 'data/KRA_PREC_100m.tif'
img = io.imread(tiff_name)
img_agr_stat = img.mean(axis = 2)
io.imsave('agregation/mean_PREC_100m.tif', img_agr_stat, plugin='tifffile')
```

```python
tiff_name = 'data/KRA_RELIEF_HEIGHT_100m.tif'
img = io.imread(tiff_name)
```

```python
plt.imshow(img)
```

```python
from scipy.ndimage import gaussian_filter
```

```python
fig, ax = plt.subplots(1, 2, figsize = (12, 8))
filt = gaussian_filter(img, 100)
ax[0].imshow(img)
ax[1].imshow(filt)
fig.axes.get_xaxis().set_visible(False)
fig.axes.get_yaxis().set_visible(False)
plt.savefig('Height.png')
```

```python
io.imsave('agregation/gausian_KRA_RELIEF_HEIGHT_100m.tif', filt, plugin='tifffile')
```
